```python
bigmacDataFrame = pd.read_csv('resources/bigmac.csv', parse_dates=['Date'], index_col=['Date', 'Country'])
```

```python
# asagidaki kod parcalarinin ikisi de outer index olan Date column'unun her satir icin degerlerini iceren bir nesne dondurur
bigmacDataFrame.index.get_level_values('Date')
bigmacDataFrame.index.get_level_values(0)
```

```python
# index column'larimin adlarini degirtirir
bigmacDataFrame.index.set_names(names=['Day', 'Location'], inplace=True)

# sadece belirtilen index column'unun adini degistirir
bigmacDataFrame.index.set_names(names='Date', level=0, inplace=True)

bigmacDataFrame.head()
```

```python
# index column'larini kendi icinde siralar, ascending=True (varsayilan) artan siralama, ascending=False azalan siralama
# iki column icin de azalan siralama yapar
bigmacDataFrame.sort_index(ascending=False)

# ilk index column icin artan siralama, ikinci index column icin azalan siralama yapar
bigmacDataFrame.sort_index(ascending=[True, False])

# belirtilen index column'una gore azalan siralama yapar
bigmacDataFrame.sort_index(level=0, ascending=False)
bigmacDataFrame.sort_index(level='Date', ascending=False)
```

```python
# multi index calisirken index olarak ilk elemani outer index degeri, ikinci elemani inner index degeri olan bir tuple kullaniriz
bigmacDataFrame.loc[('2016-01-01', 'Argentina'), "Price in US Dollars"]
bigmacDataFrame.loc[('2016-01-01', 'Argentina')]

# sadece outer index'i tek ogeli bir tuple olarak gonderebiliriz
bigmacDataFrame.loc[('2016-01-01')].head()
bigmacDataFrame.loc['2016-01-01'].head()
```

```python
# dataframe'in transpozunu alir
bigmacDataFrameTransposed = bigmacDataFrame.transpose()

# belirttigim satir ve sutundaki deger donduruldu
bigmacDataFrameTransposed.loc[('Price in US Dollars'), ('2010-01-01', 'Australia')]

# index'i belirttigim tuple'da sona virgul ekledigim icin seri donduruldu
bigmacDataFrameTransposed.loc[('Price in US Dollars', ), ('2010-01-01', 'Australia')]

# belirttigim index degeri ve column degerleri araligindaki degerleri getirir
bigmacDataFrameTransposed.loc[('Price in US Dollars',), ('2010-01-01', 'Sweden'):('2010-01-01', 'Turkey')]
```

```python
temp = bigmacDataFrame.copy()

# iki index column'lu bir multi index yapisi varsa inner index'i outer, outer index'i inner index yapar
# kalici degisiklik yapmaz
temp.swaplevel()

# degistirilecek index column'larini manuel olarak verebiliriz
temp.swaplevel('Country', 'Date').head()
```

```python
salesmenDataFrame = pd.read_csv('resources/salesmen.csv', parse_dates=['Date'])
salesmenDataFrame.head()
```

```python
# pivot olusturmak icin tekrar eden degerlerin sikliklarini kontrol ederiz
salesmenDataFrame.value_counts()
```

```python
# index olarak verilen column'u index olarak kullanir
# columns olarak verilen column'un unique degerlerini yukarida row olarak gosterir
# values olarak verilen degerleri, Salesman column'u ile Date column degerlerinin kesisimlerine yerlestirir
salesmenDataFrame.pivot(index='Date', columns='Salesman', values='Revenue').head()
```

```python
foodsDataFrame = pd.read_csv('resources/foods.csv')
```

```python
# cinsiyetlere gore ortalama yapilan odemelerin ortalamasi
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='mean')

# cinsiyetlere gore toplam yapilan odemelerin toplami
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='sum')

# cinsiyetlere gore yapilan en dusuk odeme
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='min')

# cinsiyetlere gore standart sapma belirtir
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='std')

# cinsiyetlere gore yapilan maksimum odemeleri verir
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='max')

# cinsiyetlere gore yapilan odemeleri siraladigimizda ortadaki degeri verir
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='median')

# cinsiyetlere gore yapilan harcamalarin sayisini verir
result = foodsDataFrame.pivot_table(values='Spend', index='Gender', aggfunc='count')

result 
```

```python
foodsDataFrame.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='count')
```

```python
foodsDataFrame.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='count')
pd.pivot_table(foodsDataFrame, values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='count')
```
