```python
revenueDataFrame = pd.read_csv('resources/revenue.csv', index_col='Date')
```

```python
# column'larin toplamlarinin degerler, column isimlerinin index oldugu yeni bir dataframe dondurur
revenueDataFrame.sum()
revenueDataFrame.sum(axis=0)
revenueDataFrame.sum(axis='index')
```

```python
# row'lardaki degerleri toplayarak sonuclarini ayni row'un index degeri ile yeni bir dataframe dondurur
revenueDataFrame.sum(axis=1)
revenueDataFrame.sum(axis='columns')
```

```python
# column'lara erismek icin su syntax kullanilabilinir
revenueDataFrame['Miami']
```

```python
# birden fazla column'u secmek icin su syntax kullanilabilir
revenueDataFrame[['Miami', 'New York']]
```

```python
# column'lar arasinda bulunmayan bir degeri bu sekilde eklersek butun satirlara eklenir
revenueDataFrame['Isparta'] = 500
revenueDataFrame
```

```python
# dataframe'deki veriler ile belirtilen yola yeni bir csv dosyasi olusturur
revenueDataFrame.to_csv('resources/new_revenue.csv')
```

```python
# loc parametresi ile column'larin arasindaki index'i, 
# column parametresi ile yeni column adini,
# value parametresi ile yeni column'un degerlerini veriyoruz, value parametresi olarak satir sayisina esit bir liste de verebiliriz
# allow_duplicates parametresi ile, yeni column adinda bir column'un zaten dataframe'de olmasi halinde insert isleminin yapilip yapilmayacagini belirtir
revenueDataFrame.insert(loc=4, column='Istanbul', value=650, allow_duplicates=False)

# yukaridaki kullanim ile bu kullanim aynidir
revenueDataFrame.insert(4, 'Istanbul', 650)
```

```python
# yeni column degerlerini bir series veya liste yardimiyla ekleyebiliriz
revenueDataFrame['Izmir'] = revenueDataFrame['Los Angeles'] + 25
```

```python
# bir series'in mul metodu ile butun degerlerini verilen parametre ile carpma islemine sokabiliyoruz
revenueDataFrame['Izmir'] = (revenueDataFrame['Izmir']).mul(1.25)
```

```python
# bir sutundaki ogelerin frekanslarini hesaplar
revenueDataFrame.value_counts()
```

```python
# series'in degerlerinin frekanslarini hesaplar
# ascending parametresi ile artan siralama islemi yapabiliriz
# normalize parametresi ile frekanslarin yuzdelik olarak hesaplanmasini saglayabiliriz
revenueDataFrame['Izmir'].value_counts(ascending=True, normalize=True)
```

```python
# dataframe'deki NaN degerleri ortadan kaldirmaya yarar
# axis=0 ise satir bazinda arama yapilir, axis=1 ise sutun bazinda arama yapilir, varsayilan 0'dir
# how='any', satir veya sutundaki herhangi bir deger NaN ise siler, how='all' ise satir veya sutundaki tum degerler NaN ise silme yapar
# subset ile NaN degerleri aratacagimiz column'lari belirtebiliriz
# inplace=False ise NaN degerlerinin silindigi yeni bir dataframe dondurur, inplace=True ise orijinal dataframe uzerinde degisiklik yapar
revenueDataFrame.dropna(axis=0, how='any', subset=['Izmir', 'Isparta'], inplace=False)
```

```python
# dataframe'deki NaN degerlerin arguman olarak verilen degerle degistirilmesini saglar
revenueDataFrame.fillna(0)
```

```python
# dataframe'deki NaN alanlar yerine True, geri kalan alanlar icin False degeri bulunduran bir dataframe dondurur
revenueDataFrame.isna()
```

```python
# series'de NaN eleman varsa True, yoksa False dondurur
revenueDataFrame['Izmir'].hasnans
```

```python
# verilen series'deki verilerin type'ini uygunsa arguman olarak verilen turde ayarlar
# NaN deger varsa hata verir
revenueDataFrame['Izmir'].astype('int')
```

```python
# series'deki essiz degerleri iceren bir numpy array dondurur
revenueDataFrame['Izmir'].unique()
```

```python
# series'deki essiz degerlerin sayisini verir
revenueDataFrame['Izmir'].nunique()
```

```python
# dataframe'deki verileri New York column'unun artan siralanmasina gore duzenler
# ascending=True olursa azalan sekilde siralar
revenueDataFrame.sort_values('New York', ascending=False)
revenueDataFrame.sort_values(by='New York')
```

```python
# na_position parametresini NaN iceren satirlari yeni dataframe'in neresinde gosterilecegini belirler
# first ise basta, last ise sonda gosterir. varsayilan olarak last olarak gelir
revenueDataFrame.sort_values(by='New York', na_position='first')
```

```python
# once ilk column'a gore azalan, sonra siralanan degerleri kendi icinde ikinci column'a gore artan siralar
revenueDataFrame.sort_values(by=['New York', 'Istanbul'], ascending=[True, False])
```

```python
# index'leri siralar
revenueDataFrame.sort_index()
```
