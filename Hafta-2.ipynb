```python
numbers = [1, 2, 3, 4, 5, 6.5]
numberSeries = pd.Series(numbers) # to create a new pandas series with given numbers

print(numberSeries.sum()) # gives the sum of the items in the Series
print(numberSeries.mean()) # gives the mean of the items in the Series
print(numberSeries.product()) # gives the product of the items in the Series
print(numberSeries.std()) # gives the std of the items in the Series
print(numberSeries.min()) # gives the minimum number in the Series
print(numberSeries.max()) # gives the maximum number in the series
print(numberSeries.median()) # gives the median of the Series
print(numberSeries.mode()) # gives the mode of the Series
```

```python
temp = pd.Series([1, 2, 3, 4, "test"])

len(temp) # gives the length of the given Series
temp.size # gives the length of the given Series
```

```python
temp = pd.Series([1, 2, 3, 4, None])
temp.size # it counts the None value -> 5
temp.count() # it doesn't count the None value -> 4
```

```python
temp = pd.Series([1, 2, 3, 4, "test"])
temp.is_unique # if there is no repeated item in the given Series returns True, otherwise returns False
```

```python
temp = pd.Series([1, 2, 3, 4, "test"])
temp.index # returns the index column of given Series as a RangeIndex object
temp.values # returns the data column of given Series as an numpy array
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']
indexes = ['a', 'b', 'c', 'd']

temp = pd.Series(data=days, index=indexes) # you also can set the index column of the Series
temp2 = pd.Series(days, indexes)
print(temp)
print(temp2)
```
