```python
# csv dosyasindaki Pokemon column'unu index column olarak kullanacagi bir dataframe dondurur
tempDataFrame = pd.read_csv('resources\pokemon.csv', index_col='Pokemon') 
tempDataFrame
```

```python
# indexleri indexlerine ve ilk column'u degerlerine esit olan bir series dondurur
tempSeries = tempDataFrame.squeeze('columns')
tempSeries
```

```python
# sorgulanan degerin series'in indexleri arasinda olup olmadigin gosteren bir boolean dondurur
'Bulbasaur' in tempSeries
```

```python
# tempSeries indexlerinden olusan yeni bir Index nesnesi dondurur
# yeni series'in indexleri sifirdan baslar
tempSeries.index
```

```python
# series elemanlarina ulasmak icin series[index_value] sytnax'ini kullaniriz
tempSeries['Bulbasaur']
```

```python
# ayni anda birden fazla elemana ulasmak icin series[[index_value1, index_value2, ...]] syntax'ini kullaniriz
# bu, aranan degerlerin value, arattigimiz index'lerin yine index'ler oldugu bir series dondurur
tempSeries[['Bulbasaur', 'Noibat']]
```

```python
# python listelerinde yapabildigimiz slicing islemini series'in indexlerini kullanarak series uzerinde de yapabiliriz
tempSeries['Squirtle':'Noibat']
tempSeries[:'Noibat']
tempSeries[:'Noibat':10]

# series index'leri sayisal degerli ise index'lerin arasinda olmayan sayilar kullanarak elemanlara erismeye calisirsak hata aliriz
# cunku asagidaki gibi bir yazimda python -1'i index olarak kabul edip dogal olarak bulamazsa hata verecektir
tempSeries[-1]

# slicing isleminde negatif sayilari atlama degerini belirtmedigimiz surece kullanabiliriz 
# ornegin tempSeries[10:20:-1] bos bir series dondurecektir
tempSeries[-20:-1:2]
```

```python
# series'deki bir elemana ulasmak icin get fonksiyonu ile index verebiliriz
tempSeries.get('Bulbasaur')

# series index'leri ne olursa olsun sifirdan baslayan index'ler ile de elemanlara erisebiliriz
tempSeries[0]
tempSeries.get(0)
```

```python
# su islemle aynidir: tempSeries[['Bulbasaur', 'Noibat']]
# aranilan elemanlardan bir tanesini bile bulamazsa none dondurur
tempSeries.get(['Bulbasaur', 'Noibat'])
```

```python
# get fonksiyonuna ikinci parametre olarak, eger aradimiz degeri bulamazsa dondurecegi default degeri ayarlayabiliriz
tempSeries.get('Bulbasar', 'EMPTY')
```

```python
# uygun series'ler ile aritmetik islemler yapabiliriz
# bu islemler sonucunde yapilan islemin her bir elemana uygulandigi yeni bir series dondurulur
tempSeries = pd.Series([1, 2, 3, 4, 5])

tempSeries + 10
tempSeries * 10
tempSeries / 5
```

```python
# series elemanlarina 5 ekler ve bunu orijinal series uzerinde yapar
tempSeries.add(5)
```

```python
# her bir degerden kac tane oldugunu gosteren bir series dondurur
# index degerleri, asil series'deki degerler ve value degerleri ise o series'de bunlardan kacar tane oldugu gosteren sayi olur
pokemonColumnSeries.value_counts()
```

```python
# normalize parametresini True verirsek bu sefer elemanlardan kacar tane oldugunu degil,
# elemanlarin birbirlerinin sayisina gore oranini 0 - 1 araliginda verir
pokemonColumnSeries.value_counts(normalize=True)

# bunu 100 ile carparsak hangi degerden kacar tane oldununun yuzdelik oranini buluruz
pokemonColumnSeries.value_counts(normalize=True) * 100
```

```python
# dondurulen series'deki index'lerin siralanmasi icin sort parametresi True, orijinal sirasinda goruntulemek icin ise False veriririz
pokemonColumnSeries.value_counts(sort=True)
```

```python
def addOne(value):
    return value + 1

# apply ile verilen fonksiyon, tum elemanlara uygulanarak degistirilmis elemanlardan olusan yeni bir series dondurur
tempSeries.apply(addOne)
```

```python
def function(value):
    if('a' in value):
        return 'it has a letter'
    elif('b' in value):
        return 'it has b letter'
    else:
        return 'it doesn\'t have a or b letter'

# map metodu ile her bir elemana bir fonksiyon uygulayabiliriz
pokemonColumnSeries.map(function)
```

```python
mapping = {
    1 : 'bir',
    2 : 'iki',
    3 : 'uc',
    4 : 'dort',
    5 : 'bes'
}
# map fonksiyonu ile daha onceden tanimlagidimiz mapping'i uygulayabiliriz
# herhangi bir deger icin mapping belirtmezsek degerlerin hepsi NaN olacaktir
tempSeries.map(mapping)
```
