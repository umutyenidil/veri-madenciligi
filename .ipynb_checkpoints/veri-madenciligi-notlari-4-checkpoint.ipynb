```python
import pandas as pd
```

```python
bondDataFrame = pd.read_csv('resources/jamesbond.csv', index_col='Film')
bondDataFrame.sort_index(inplace=True)
bondDataFrame.head()
```

```python
# Actor sutunundaki degerleri 'Sean Connery'a esit olan satirlar icin True geri kalan satirlar icin False'a karsilik gelen
# boolean degerler iceren bir series dondurur
condition = bondDataFrame['Actor'] == 'Sean Connery'

# asagidaki iki satir, ayni sonucu uretir
bondDataFrame[condition]['Actor']
bondDataFrame.loc[condition, 'Actor'].head()
```

```python
# index degerlerini veya column adlarin degistirdigimiz bir dataframe dondurur:
# mapper, degistirilecek degerin key, yeni degerin value oldugu bir sozluk alir
# axis ise degistirilecek degerin index degeri mi yoksa column adi mi oldugunu belirtir, 0, index, 1, column degerlerini alabilir
# inplace, True verilirse degisiklik orijinal dataframe uzerinde yapilir False verilirse yeni bir dataframe dondurur
bondDataFrame.rename(mapper={
    'GoldenEye' : 'Golden Eye',
    'The World Is Not Enough' : 'Best Bond Movie Ever'
}, axis=0, inplace=False)

# bu kullanimda axis ve mapper parametrelerini tek bir sekilde verebiliyoruz
# index parametresine degisiklik yapilmasini istedigimiz index degerlerini key, yeni degerleri value olan bir sozluk veriyoruz
# column parametresine degisiklik yapilmasini istedigimiz column adlarini key, yeni adlari value olan bir sozluk veriyoruz
# ayni isimli birden fazla column veya ayni index'ten birden fazla varsa degisikligi butun column/index'ler uzerinde yapar
bondDataFrame.rename(index={
  'Casino Royale' : 'Casino Royale Updated'
}, inplace=False).head()
```

```python
# dataframe'in column isimlerini iceren bir index nesnesi dondurur
bondDataFrame.columns

temp = bondDataFrame.copy()

# column degerlerini kalici olarak degistirebiliriz 
temp.columns = ['Yil', 'Basrol', 'Yonetmen', 'Box Office', 'Butce(M$)', 'Basrol Maasi']
temp.head()
```

```python
temp = bondDataFrame.copy()

# verilen index degerine denk gelen satiri silinmis yeni bir dataframe dondurur
# dataframe'de bulunmayan bir deger verilirsa hata alinir
# varsayilan olarak axis=0 yani index bazinda silme yapar, axis=1 dataframe'den column'u siler
# inplace parametresi mevcuttur
result = temp.drop('Casino Royale', inplace=False, axis=0)
result = temp.drop(['Casino Royale', 'Dr. No'], inplace=False)
result = temp.drop(['Box Office'], inplace=False, axis=1)

result.head()
```

```python
temp = bondDataFrame.copy()

# degeri verilen column'u dataframe'den kalici olarak siler ve silinen column'u series olarak geri dondurur
result = temp.pop('Actor')

result.head()
```

```python
temp = bondDataFrame.copy()

# belirtilen column'u kalici olarak siler
del temp['Actor']

temp.head()
```

```python
# rastgele n sayida satir getirir
result = bondDataFrame.sample(n=1)

# dataframe'in satir sayisindan, belirtilen oran kadarini rastgele getirir
result = bondDataFrame.sample(frac=0.25)

# rastgele n sayidaki column'u getirir
result = bondDataFrame.sample(n=2, axis=1)

result.head()
```

```python
# verilen column'daki en buyuk n degeri bir dataframe ile dondurur
# verilen column'daki degerlere gore buyukten kucuge dogru siralanmistir
bondDataFrame.nlargest(3, 'Box Office')
```

```python
# verilen column'daki en kucuk n degeri bir dataframe ile dondurur
# verilen column'daki degerlere gore kucukten buyuge dogru siralanmistir
bondDataFrame.nsmallest(3, 'Box Office')
```

```python
# dataframe'de belirtilen kosula uyan satirlari aynen tutarken, uyumayan satirlardaki verileri NaN ile degistirilmis
# yeni bir dataframe dondurur
result = bondDataFrame.where(bondDataFrame['Actor']=='Roger Moore')
result = bondDataFrame.where((bondDataFrame['Actor']=='Sean Connery') & (bondDataFrame['Bond Actor Salary']>3))

# dropna ile NaN verilerin bulundugu satirlari silerek dataframe'i temizleyebiliriz
result = result.dropna()

result
```

```python
temp = bondDataFrame.copy()

# query metodu ile string olarak dataframe icin sorgu yazabiliyoruz ama column adinda bosluk varsa hata aliriz
result = temp.query("Actor == 'Roger Moore'")
result = temp.query("Year > 600")

# column adlarindaki bosluklari su sekilde ortadan kaldirabiliriz
temp.columns = [columnName.replace(' ', '_') for columnName in temp.columns]

result = temp.query('Box_Office > 600')
result = temp.query("Actor == 'Roger Moore' and Director=='John Glen'")
result = temp.query("Actor in ['Timothy Dalton', 'George Lazenby']")



result.head()
```

```python
addCurrency = lambda text : str(text) + ' Million $'

# apply, kendisine verilen tek parametreli fonksiyonu series'in butun elemanlarina uygulayip, degisen elemanlardan olusan
# yeni bir series dondurur
bondDataFrame['Bond Actor Salary'].apply(addCurrency).head()
```

```python
def doFilter(row):
    if(row[1] == 'Sean Connery'):
        return 'anladim'
    return 'test'

# axis=0 ile
# her column'u birer series olarak verilen fonksiyona gecer
# yani apply ile bir kere cagiririz ama column sayisi kadar calisir bu fonksiyon
# her seferinde de bir column'u series halinda arguman olarak fonksiyona gonderir
bondDataFrame.apply(doFilter, axis=0)
```

```python
chicagoDataFrame = pd.read_csv('resources/chicago.csv')

chicagoDataFrame.head()
```

```python
# verilen column'daki degerlerin ilk harflerini buyuk, gerisini kucuk yaparak ilgili column'u yeni bir series olarak dondurur
result = chicagoDataFrame['Department'].str.title()

# verilen column'daki degerlerin butun harflerini kucuk yaparak ilgili column'u yeni bir series olarak dondurur
result = chicagoDataFrame['Department'].str.lower()

# verilen column'daki degerlerin karakter sayilarini yeni bir series olarak dondurur
result = chicagoDataFrame['Department'].str.len()

# verilen column'daki degerlerdeki belirtilen kisimlari degistirerek ilgili column'u yeni bir series olarak dondurur
# regex kullanip kullanamayacagimizi belirtebiliyoruz
result = chicagoDataFrame['Department'].str.replace('MGMNT', 'MANAGEMENT', regex=False)

# verilen column'daki degerlerin verilen karakterlerle baslayip baslamadigina gore bir bool series dondurur
result = chicagoDataFrame['Department'].str.startswith('W')

# verilen column'daki degerlerin bastan ve sondan verilen karakterleri varsa silinmis yeni bir series dondurur
# arguman verilmezse bosluklari siler
result = chicagoDataFrame['Department'].str.strip('W')


# verilen column'lardaki degerlerin her birini verilen karakterlerden bolerek, bolunmus parcalari liste olarak dondurur
result = chicagoDataFrame['Name'].str.split(',')

# verilen listelerin sifirnci indexli elemanlarini iceren bir series dondurur
result = chicagoDataFrame['Name'].str.split(',').str.get(0)

result.head()
```
