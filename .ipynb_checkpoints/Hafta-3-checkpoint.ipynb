```python
dataframe = pd.read_csv('resources/employees.csv');

dataframe['Start Date'] = pd.to_datetime(dataframe['Start Date'])
dataframe['Last Login Time'] = pd.to_datetime(dataframe['Last Login Time'])
dataframe['Senior Management'] = dataframe['Senior Management'].astype('bool')
dataframe['Gender'] = dataframe['Gender'].astype('category')

dataframe.info()
```

```python
dataframe = pd.read_csv('resources/employees.csv', parse_dates=['Start Date', 'Last Login Time']);
dataframe['Senior Management'] = dataframe['Senior Management'].astype('bool')
dataframe['Gender'] = dataframe['Gender'].astype('category')

dataframe.info()
```

```python
dataframe = pd.read_csv('resources/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
dataframe['Senior Management'] = dataframe['Senior Management'].astype('bool')
dataframe['Gender'] = dataframe['Gender'].astype('category')

dataframe[dataframe['Gender'] == 'Male'] # gives the rows which ones has Male value on their Gender column
```
