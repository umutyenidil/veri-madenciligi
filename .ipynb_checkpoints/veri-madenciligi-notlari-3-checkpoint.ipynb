```python
# employeesDataFrame'deki Gender column degeri 'Male' olan alanlar icin True, digerleri icin False olan bir series dondurur
condition = employeesDataFrame['Gender'] == 'Male'

# Team column degeri Finance olan alanlar icin True, digerleri icin False dondurur
condition = emplooyesDataFrame['Team'] == 'Finance'

# Senior Management column'u bool degerler icerir, yani yukarida yaptigimiz sorgulama islemlerinden donen boolean degerler
# iceren series'ler gibi bir series'dir
condition = employeesDataFrame['Senior Management']

# Team column degeri Finance olmayan alanlar icin True, digerleri icin False dondurur
condition = employeesDataFrame['Team'] != 'Finance'

# Salary column degeri 100000'den buyuk olan satirlar icin True, digerleri icin False denk gelecek bir series dondurur
condition = employeesDataFrame['Salary'] >= 100000

# Start Date column degeri belirtilen tarihten yeni olan tarihlerin bulundugu alanlar icin True
# digerleri icin False deger iceren bir series dondurur
condition = employeesDataFrame['Start Date'] > '1990-01-01'

# Gender column degeri 'Male' ve Team column degeri 'Finance' olan alanlar icin True
# digerleri icin False degerler iceren bir series dondurur
# ampersent yerine python'un 've' mantiksal operatoru olan 'and' kullanirsak hata aliriz 
condition = (employeesDataFrame['Gender'] == 'Male') & (employeesDataFrame['Team'] == 'Finance')

# Team column degeri 'Marketing' veya Start Date column degeri '1985-01-01' olan alanlar icin True
# digerleri icin False degerler iceren bir series dondurur
# dikey cizgi yerine python'un 'veya' mantiksal operatoru olan 'or' kullanirsak hata aliriz 
condition = (employeesDataFrame['Team'] == 'Marketing') | (employeesDataFrame['Start Date'] == '1985-01-01')

# Team column degeri 'Marketing' veya Start Date column degeri '1985-01-01' olan ve 
# First Name column degeri 'Robert' alanlar icin True digerleri icin False degerler iceren bir series dondurur
condition = ((employeesDataFrame['Team'] == 'Marketing') | (employeesDataFrame['Start Date'] == '1985-01-01')) & (employeesDataFrame['First Name'] == 'Robert')

# condition'daki True olan alanlarla eslesen satirlari getirir
employeesDataFrame[condition].head()
```

```python
# bu condition ile altindaki condition ayni sonucu uretir
condition = (employeesDataFrame['Team'] == 'Marketing') | (employeesDataFrame['Team'] == 'Product') | (employeesDataFrame['Team'] == 'Finance')

# isin metodu, kendisine verilen kume icindeki degerlerin oldugu satirlar icin True digerleri icin False degerler iceren bir
# series dondurur ve bunu elimizdeki dataframe'i filtrelemek icin kullanabiliriz
condition = employeesDataFrame['Team'].isin({'Marketing', 'Product', 'Finance'})

employeesDataFrame[condition].head()
```

```python
# Team column'unda degeri null olan degerler icin True, digerleri icin False degerler iceren bir series dondurur
condition = employeesDataFrame['Team'].isnull()

employeesDataFrame[condition].head()
```

```python
# Team column'unda degeri null olmayan degerler icin True, digerleri icin False degerler iceren bir series dondurur
condition = employeesDataFrame['Team'].notnull()

employeesDataFrame[condition].head()
```

```python
# Salary column degeri 40000 ile 60000 arasinda olan degerler icin True, digerleri icin False degerler iceren bir series verir
# inclusive parametresi, verilen degerlerin dahil olup olmadigini belirtir, True ise degerler dahildir False ise degildir
# inclusive parametresinin varsayilan degeri True'dur
condition = employeesDataFrame['Salary'].between(40000, 60000, inclusive=True)

employeesDataFrame[condition].head()
```

```python
# Start Date column degeri belirtilen tarihler arasinda olan degerler icin True, digerleri icin False degerler iceren bir
# series dondurur
condition = employeesDataFrame["Start Date"].between("1985-01-01", "1995-01-01")

employeesDataFrame[condition].head()
```

```python
# Last Login Time column degeri belirtilen saatlar arasinda olan degerler icin True, digerler icin False degerler iceren 
# bir series dondurur
# inclusive=False ile inclusive='neither' ayni seydir
condition = employeesDataFrame["Last Login Time"].between("08:30 AM", "12:30 PM", inclusive="neither")

employeesDataFrame[condition].head()
```

```python
# First Name column'undaki degerlerden tekrar eden varsa, ilk buldugu degere False sonradan ilk degerle ayni olan butun
# satirlara True deger atanmis bir series dondurur
# keep parametresini False olarak ayarlarsak, ilk bulunan degerden sonraki butun ayni degerlere True verir series icinde
# keep parametresini 'first' olarak ayarlarsak ilk bulunan degerden sonraki butun ayni degerlere True verir series icinde
# keep parametresini 'last' olarak ayarlarsak, tekrar eden degerlerden index olarak en buyuk olaninin (yani en sondaki)
# karsiliginda False, onceki tekrar eden degerlerin hepsine True atanacak bir series dondurur
condition = employeesDataFrame['First Name'].duplicated(keep=False)

employeesDataFrame[condition].head()
```

```python
# verilen series'deki boolean degerlerin degerini alabiliriz ~ operatoru ile
condition = [~condition]
```

```python
# tekrar eden satirlardan ilkini tutup diger satirlari silinmis yeni bir dataframe dondurur
employeesDataFrame.drop_duplicates().head()

# tekrar eden degerleri aramak istedigimiz column'lari subset parametresi ile belirtebiliyoruz
# verilen column'lardaki tekrar eden deger iceren satirlarin ilkini tutup diger satirlari silinmis bir dataframe dondurur
employeesDataFrame.drop_duplicates(subset=['First Name', 'Team']).head()
```

```python
# verilen series icindeki degerlerin her birinden bir tane iceren bir numpy array dondurur
# NaN degeri de bir deger sayar
employeesDataFrame['First Name'].unique()[0:5]
```

```python
# verilen series icindeki degerlerin her birinden bir tanesini sayacak sekilde deger dondurur
# NaN degeri saymaz varsayilan olarak
# dropna=False verilirse NaN'i da sayar
employeesDataFrame['First Name'].nunique(dropna=False)
```

```python
# asagidaki iki satirda da ayni islemi yaptik, ilki zaten olan bir dataframe'in index column'unu istedigimiz column ile
# degistirmemizi saglar
# ikinci column'da ise dataframe'i ice aktarirken index olarak ayarlamak istedigimiz column'u index_col parametresi
# ile belirtiriz
bondDataFrame.set_index(keys='Actor', inplace=True)
bondDataFrameWithFilmIndexColumn = pd.read_csv('resources/jamesbond.csv', index_col='Film')
```

```python
# degistirdigimiz index_column'u orijinal haline getirir, drop=True yaparsak eski index column'u dataframe'den silinir
# inplace False ise degisikligin uygulandigi yeni bir dataframe dondurur, True ise degisikligi orijinal dataframe'e uygular
bondDataFrame.reset_index(drop=False, inplace=False).head()
```

```python
temp = bondDataFrame.reset_index()
bondDataFrame2 = temp.drop_duplicates(subset='Actor')
bondDataFrame2.set_index(keys='Actor', inplace=True)
bondDataFrame2.head()
```

```python
# verilen string turundeki index degerinin bulundugu satirlari, index olarak aldigi column adlarini bir series olarak dondurur
bondDataFrame2.loc['George Lazenby']
```

```python
# verilen index'lerdeki satir araligini iceren bir dataframe dondurur
# slicing islemini python'daki gibi yapabiliriz
bondDataFrame2.loc['David Niven':'Roger Moore']
bondDataFrame2.loc['David Niven':'Roger Moore':2]
```

```python
# sectigimiz column'lari dataframe olarak dondurur
bondDataFrame2.loc[["David Niven", "Roger Moore", 'George Lazenby']]
```

```python
# loc attribute ile yaptigimiz islemleri iloc ile sayisal index bazinda yapabiliriz
bondDataFrame.iloc[1:3]
```

```python
# iloc veya loc attribute'unda koseli parantez icine yazilan ve virgulle ayrilan iki degerden
# ilki, belirtilen index'leri secer. burada liste olarak istenen index'leri verebilir veya slicing yapabiliriz
# ikincisi, belirtilen column'lari secer. burada liste olarak istenen column'lar verilebilir veya slicing yapilabilir
# iloc kullanirsak index ve column'larin sayisal index'lerini vermek durumundayiz
# loc kullanirsak index ve column'larin string degerlerini vererek islem yapariz
bondDataFrame.iloc[[0, 2, 4], [1, 3, 5]]
```
