```python
# bu condition ile altindaki condition ayni sonucu uretir
condition = (employeesDataFrame['Team'] == 'Marketing') | (employeesDataFrame['Team'] == 'Product') | (employeesDataFrame['Team'] == 'Finance')

# isin metodu, kendisine verilen kume icindeki degerlerin oldugu satirlar icin True digerleri icin False degerler iceren bir
# series dondurur ve bunu elimizdeki dataframe'i filtrelemek icin kullanabiliriz
condition = employeesDataFrame['Team'].isin({'Marketing', 'Product', 'Finance'})

employeesDataFrame[condition].head()
```

```python
# Team column'unda degeri null olan degerler icin True, digerleri icin False degerler iceren bir series dondurur
condition = employeesDataFrame['Team'].isnull()

employeesDataFrame[condition].head()
```

```python
# Team column'unda degeri null olmayan degerler icin True, digerleri icin False degerler iceren bir series dondurur
condition = employeesDataFrame['Team'].notnull()

employeesDataFrame[condition].head()
```

```python
# Salary column degeri 40000 ile 60000 arasinda olan degerler icin True, digerleri icin False degerler iceren bir series verir
# inclusive parametresi, verilen degerlerin dahil olup olmadigini belirtir, True ise degerler dahildir False ise degildir
# inclusive parametresinin varsayilan degeri True'dur
condition = employeesDataFrame['Salary'].between(40000, 60000, inclusive=True)

employeesDataFrame[condition].head()
```

```python
# Start Date column degeri belirtilen tarihler arasinda olan degerler icin True, digerleri icin False degerler iceren bir
# series dondurur
condition = employeesDataFrame["Start Date"].between("1985-01-01", "1995-01-01")

employeesDataFrame[condition].head()
```

```python
# Last Login Time column degeri belirtilen saatlar arasinda olan degerler icin True, digerler icin False degerler iceren 
# bir series dondurur
# inclusive=False ile inclusive='neither' ayni seydir
condition = employeesDataFrame["Last Login Time"].between("08:30 AM", "12:30 PM", inclusive="neither")

employeesDataFrame[condition].head()
```
